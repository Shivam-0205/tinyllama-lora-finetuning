```python
from datasets import load_dataset
tinystories = load_dataset(
    "roneneldan/TinyStories",
    split="train[:1000]"
)
print(tinystories)
```

```python
wiki = load_dataset(
    "Salesforce/wikitext",
    "wikitext-2-raw-v1",
    split="train[:1000]"
)
print(wiki[0])
```

```python
agnews = load_dataset(
    "ag_news",
    split="train[:1000]"
)

print(agnews[0])
```

```python
def normalize_tinystories(example):
    return {
        "text": example["text"]
    }
def normalize_agnews(example):
    return {
        "text": example["text"]
    }
def normalize_wiki(example):
    return {
        "text": example["text"]
    }
```

```python
tiny_clean = tinystories.map(normalize_tinystories)
wiki_clean = wiki.map(normalize_wiki)
agnews_clean = agnews.map(normalize_agnews)
```

```python
print(tiny_clean[0])
print(wiki_clean[0])
print(agnews_clean[0])
```

```python
wiki_clean = wiki_clean.filter(
    lambda x: len(x["text"].strip()) > 0
)
```

```python
print(tiny_clean[0])
print(wiki_clean[0])
print(agnews_clean[0])
```

```python
from datasets import concatenate_datasets
combined_dataset = concatenate_datasets([
    tiny_clean,
    wiki_clean,
    agnews_clean
])
print(combined_dataset)
```

```python
combined_dataset = combined_dataset.remove_columns(
    ["label"]
)
print(combined_dataset)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(
    "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)
tokenizer.pad_token = tokenizer.eos_token
```

```python
def tokenize_function(example):
    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )
```

```python
tokenized_dataset = combined_dataset.map(
    tokenize_function,
    batched=True
)
```

```python
print(tokenized_dataset[0])
```

```python
from datasets import load_dataset
yelp = load_dataset(
    "yelp_review_full",
    split="train[:500]"
)
print(yelp[0])
```

```python
xsum = load_dataset(
    "xsum",
    split="train[:500]"
)

print(xsum[0])
```

```python
def normalize_yelp(example):
    return {"text": example["text"]}
def normalize_xsum(example):
    return {"text": example["document"]}
```

```python
yelp_clean = yelp.map(normalize_yelp)
xsum_clean = xsum.map(normalize_xsum)
```

```python
yelp_clean = yelp_clean.remove_columns(
    [col for col in yelp_clean.column_names if col != "text"]
)
xsum_clean = xsum_clean.remove_columns(
    [col for col in xsum_clean.column_names if col != "text"]
)
```

```python
from datasets import concatenate_datasets

combined_dataset = concatenate_datasets([
    tiny_clean,
    wiki_clean,
    agnews_clean,
    yelp_clean,
    xsum_clean
])
```

```python
from transformers import AutoModelForCausalLM, BitsAndBytesConfig
import torch

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)

model = AutoModelForCausalLM.from_pretrained(
    "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    quantization_config=bnb_config,
    device_map="auto"
)

print("Model loaded in 4-bit successfully!")
```

```python
from peft import prepare_model_for_kbit_training
model = prepare_model_for_kbit_training(model)
print("Model prepared for LoRA training!")
```

```python
from peft import LoraConfig
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)
print("LoRA config created!")
```

```python
from peft import get_peft_model
model = get_peft_model(
    model,
    lora_config
)
model.print_trainable_parameters()
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./outputs",

    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,

    learning_rate=2e-4,
    num_train_epochs=1,

    logging_steps=10,
    save_steps=100,

    save_total_limit=1,

    fp16=True,

    optim="paged_adamw_8bit",

    report_to="wandb",

    logging_dir="./logs",

    warmup_ratio=0.03,

    lr_scheduler_type="cosine",

    push_to_hub=False
)

print("Training arguments ready!")
```

```python
from transformers import DataCollatorForLanguageModeling
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)
```

```python
from transformers import Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)
print("Trainer created successfully!")
```

```python
import wandb
wandb.init(
    project="tinyllama-lora-project",
    name="tinyllama-lora-run"
)
```

```python
trainer.train()
```

```python
model.save_pretrained("./tinyllama-lora-model")
tokenizer.save_pretrained("./tinyllama-lora-model")
print("Model saved successfully!")
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=50
)

prompt = "Artificial intelligence is"
result = pipe(prompt)
print(result[0]["generated_text"])
```

```python
import time

prompt = "Machine learning is"

start_time = time.time()

result = pipe(prompt)

end_time = time.time()

generated_text = result[0]["generated_text"]

print(generated_text)

print("\nInference Time:", end_time - start_time, "seconds")
```

```python
num_tokens = len(
    tokenizer.encode(generated_text)
)

tokens_per_second = num_tokens / (end_time - start_time)
print("Tokens generated:", num_tokens)
print("Tokens/sec:", tokens_per_second)
```
